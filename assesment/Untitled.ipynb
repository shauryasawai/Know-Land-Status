```python
data = {
    "District1": {
        "Mandal1": {
            "Village1": ["Survey1", "Survey2", "Survey3"],
            "Village2": ["Survey4", "Survey5"]
        },
        "Mandal2": {
            "Village3": ["Survey6", "Survey7"],
            "Village4": ["Survey8"]
        }
    },
    "District2": {
        "Mandal3": {
            "Village5": ["Survey9", "Survey10"],
            "Village6": ["Survey11", "Survey12"]
        },
        "Mandal4": {
            "Village7": ["Survey13"],
            "Village8": ["Survey14", "Survey15", "Survey16"]
        }
    },
    "District3": {
        "Mandal5": {
            "Village9": ["Survey17", "Survey18"],
            "Village10": ["Survey19", "Survey20"]
        },
        "Mandal6": {
            "Village11": ["Survey21"],
            "Village12": ["Survey22", "Survey23", "Survey24"]
}
    },
    "District4": {
        "Mandal7": {
            "Village13": ["Survey25", "Survey26"],
            "Village14": ["Survey28", "Survey29"]
        },
        "Mandal8": {
            "Village15": ["Survey30"],
            "Village16": ["Survey31", "Survey32", "Survey33"]
}
    }
}


def get_survey_numbers(district, mandal, village):
    if district in data and mandal in data[district] and village in data[district][mandal]:
        return data[district][mandal][village]
    else:
        return []

district = input("Please Enter Your District")
mandal = input("Please Enter Your Mandal")
village = input("Please Enter Your Village")

survey_numbers = get_survey_numbers(district, mandal, village)
if survey_numbers:
    print("Survey numbers for {} - {} - {}:".format(district, mandal, village))
    for survey in survey_numbers:
        print(survey)
else:
    print("No survey numbers found for the specified district, mandal, and village.")
```
